## **Data Acquisition through OpenStreetMap (OSM) API**

```python
import requests
import csv
import pandas as pd

# for shp files
import geopandas as gpd
import os
from shapely.geometry import shape
```

```python

# Define the Overpass API URL
overpass_url = "http://overpass-api.de/api/interpreter"

# Define the Overpass QL query for New York
query = """
[out:json][timeout:25];
(
  node["amenity"="park"](40.477399, -74.25909, 40.917577, -73.700272);
);
out body;
"""

# Send the request
response = requests.get(overpass_url, params={"data": query})
data = response.json()

#  CSV file to write
csv_file = "new_york_amenities-park.csv"
with open(csv_file, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    
    #  header
    writer.writerow(["id", "lat", "lon", "tags"])
    
    #  rows
    for element in data["elements"]:
        
        element_id = element.get("id")
        lat = element.get("lat")
        lon = element.get("lon")
        tags = element.get("tags", {})
        
        
        tags_str = "; ".join(f"{k}: {v}" for k, v in tags.items())
        
        
        writer.writerow([element_id, lat, lon, tags_str])

print(f"Data successfully saved to {csv_file}")
```
