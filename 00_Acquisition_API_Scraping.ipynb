## **Data Acquisition through Web Scraping**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.walkscore.com/score/%7Bneighborhood.replace('%20',%20'-')%7D-new-york-ny#"
# Send a GET request to the website
response = requests.get(url)
response
```

```python

neighborhoods = [
   'New York, NY 10022', 'New York, NY 10019', 'New York, NY 10065',
       'New York, NY 10027', 'New York, NY 10014', 'New York, NY 10013',
       'New York, NY 10002', 'New York, NY 10075', 'New York, NY 10128',
       'New York, NY 10012', 'New York, NY 10025', 'New York, NY 10023',
       'New York, NY 10029', 'New York, NY 10031', 'New York, NY 10021',
       'New York, NY 10032', 'New York, NY 10007', 'New York, NY 10024',
       'New York, NY 10016', 'New York, NY 10036', 'New York, NY 10028',
       'New York, NY 10017', 'New York, NY 10033', 'New York, NY 10309',
       'New York, NY 10001', 'New York, NY 10003', 'New York, NY 10069',
       'New York, NY 10463', 'New York, NY 10009', 'New York, NY 10030',
       'New York, NY 10011', 'New York, NY 10026', 'New York, NY 10034',
       'New York, NY 10005', 'New York, NY 10004', 'New York, NY 10035',
       'New York, NY 10044', 'New York, NY 10037', 'New York, NY 10018',
       'New York, NY 10010', 'New York, NY 10040', 'New York, NY 10038',
       'New York, NY 11223', 'New York, NY 10314', 'New York, NY 11101',
       'New York, NY 10280', 'New York, NY 10039', 'New York, NY 10006',
       'New York, NY 10282'
]


base_url = "https://www.walkscore.com/score/"
all_data = []

# Loop
for neighborhood in neighborhoods:
    formatted_neighborhood = neighborhood.replace(", ", "-").replace(" ", "-")
    url = f"{base_url}{formatted_neighborhood}"

    try:
        response = requests.get(url)
        response.raise_for_status()
        print(f"Successfully fetched the webpage for {neighborhood}!")

        # Parse the webpage content
        soup = BeautifulSoup(response.content, "html.parser")

        # Extract the score titles
        titles = []
        badge_divs = soup.find_all("div", class_="block-header-badge score-info-link")
        for div in badge_divs:
            img_tag = div.find("img")
            if img_tag and 'alt' in img_tag.attrs:
                score_info = img_tag['alt']
                titles.append(score_info)

        # Extract location details
        location_tag = soup.find("p", class_="tight-top tight-bot clear-smallphone dull-link unstyled")
        location = location_tag.text.strip() if location_tag else neighborhood

        # dictionary
        data = {
            "Title": titles,
            "Location": location
        }

        # Convert to a DataFrame
        df = pd.DataFrame(data)
        df_new = df['Title'].str.extract(r'(\d+)\s+(.*?)\sof this location')
        df_new.columns = ['Score', 'Description']

        # Pivot and add neighborhood details
        df_pivoted = df_new.pivot_table(index=None, columns='Description', values='Score', aggfunc='first')
        df_pivoted = df_pivoted.reset_index(drop=True)
        df_pivoted["Location"] = location

        # Append to the main list
        all_data.append(df_pivoted)

    except Exception as e:
        print(f"Failed to fetch data for {neighborhood}: {e}")


NY_df = pd.concat(all_data, ignore_index=True)

NY_df['Neighborhood'] = neighborhoods
NY_df
```

```python

# List of URLs to process
urls = [
    "https://www.walkscore.com/score/New York, NY 10022",
    "https://www.walkscore.com/score/New York, NY 10013",
    "https://www.walkscore.com/score/New York, NY 10075",
    "https://www.walkscore.com/score/New York, NY 10001",
    "https://www.walkscore.com/score/New York, NY 10003",
    "https://www.walkscore.com/score/New York, NY 10005",
    "https://www.walkscore.com/score/New York, NY 10038",
    "https://www.walkscore.com/score/New York, NY 10006"

]


scores = {}

# Loop through each URL
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    location_name = url.split("/score/")[-1].replace(",", "")


    badge_divs = soup.find_all("div", class_="block-header-badge score-info-link")
    titles = []

    # Loop through each div to extract the scores
    for div in badge_divs:
        img_tag = div.find("img")
        if img_tag and 'alt' in img_tag.attrs:
            score_info = img_tag['alt']
            titles.append(score_info)

    # Store scores in the dictionary
    scores[location_name] = titles

# Print the results
for location, score_list in scores.items():
    print(f"{location}: {', '.join(score_list)}")
```
